```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
import AO3
import time 
```

```python
base_ao3 = 'https://archiveofourown.org/tags/Mahou%20Shoujo%20Madoka%20Magika%20%7C%20Puella%20Magi%20Madoka%20Magica/works'
```

```python
names = []
authors = []
ratings = []
dates = []
work_ids = []
for i in range(210):
    url = base_ao3 + '?page=' + str(i+1)
    print(url)
    fics_req = requests.get(url).text
    fics_soup = BeautifulSoup(fics_req)
    ids = fics_soup.select('ol.work > li > div > h4 > a[href]:first-of-type')
    for id in ids:
        id = id['href']
        id = id.replace('/works/','')
        work_ids.append(id)
        print(id)
    for n in fics_soup.select('ol.work > li > div > h4 > a:not([rel=author]):first-of-type'):
        names.append(n.text)
    for a in fics_soup.select ('ol.work > li > div > h4 > a[rel=author]'):
        authors.append(a.text)
    for r in fics_soup.select('ol.work > li > div > ul.required-tags > li:first-of-type > a > span:first-of-type'):
        ratings.append(r.text)
    for d in fics_soup.select('ol.work > li > div > p'):
        dates.append(d.text)
    sleep(5)
```

```python
rows_tups = list(zip(names, authors, ratings, dates))
```

```python
df = pd.DataFrame(rows_tups, columns=['names', 'authors', 'ratings', 'u_date'])
```

```python
df.to_csv('madoka_ao3.csv', encoding='utf-8-sig')
```
