```python
base_url = 'https://www.fanfiction.net/anime/Puella-Magi-Madoka-Magica-%E9%AD%94%E6%B3%95%E5%B0%91%E5%A5%B3%E3%81%BE%E3%81%A9%E3%81%8B-%E3%83%9E%E3%82%AE%E3%82%AB/?&srt=1&r=10'
```

```python
driver = uc.Chrome(use_subprocess=True)
```

```python
names = []
dates = []
authors =[]
ratings = []
languages = []
for i in range(1, 111):
    page_url = base_url + '&p=' + str(i+1)
    page_sel = driver.get(page_url)
    names_sel = driver.find_elements(By.CSS_SELECTOR, "div > a.stitle")
    for name in names_sel:
        names.append(name.text)
    p_dates_sel = driver.find_elements(By.XPATH, "//div[contains (@class, 'xgray')]/text()[contains(.,'Published')]//following-sibling::span")
    for date in p_dates_sel:
        date = date.get_attribute("data-xutime")
        date = datetime.datetime.fromtimestamp(int(date)).strftime('%Y-%m-%d') 
        dates.append(date)
    tags = driver.find_elements(By.XPATH, "//div[contains (@class, 'xgray')]")
    for tag in tags:
        tag = tag.text
        ratings.append(re.search(r'Rated: (.*?) -', tag).group(1))
        languages.append(re.search(r'Rated: (.*?) - (.*?) -', tag).group(2))
    bios = driver.find_elements(By.XPATH, "//div[contains (@class, 'z-list') and contains (@ class, 'zhover') and contains (@ class, 'zpointer')]/text()[contains(.,'by')]//following-sibling::a[1]")
    for i, bio in enumerate(bios):
        bio = bio.text
        authors.append(bio)
    sleep(5)
    
```
